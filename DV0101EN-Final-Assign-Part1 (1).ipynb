### TASK 1.1: Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year


<details><summary>Click here for a hint</summary>

<p>
You will require:-
<br>to group the year and calculate the average on the 'Automobile Sales', as the data has years and months column
<br>make use of .plot() with kind = 'line'
<br>donot forget to include labels and title
</details>


````python
# Group the data by year and calculate the total sales for each year
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum().reset_index()

# Create the line chart
plt.figure(figsize=(12, 6))
yearly_sales.plot(x='Year', y='Automobile_Sales', kind='line', ax=plt.gca())

# Customize the chart
plt.title('Automobile Sales Fluctuation Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Automobile Sales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add markers to the data points
plt.scatter(yearly_sales['Year'], yearly_sales['Automobile_Sales'], color='red', s=50)

# Annotate the highest and lowest points
max_sales = yearly_sales.loc[yearly_sales['Automobile_Sales'].idxmax()]
min_sales = yearly_sales.loc[yearly_sales['Automobile_Sales'].idxmin()]

plt.annotate(f'Highest: {max_sales["Automobile_Sales"]:.0f}', 
             xy=(max_sales['Year'], max_sales['Automobile_Sales']),
             xytext=(10, 10), textcoords='offset points', ha='left', va='bottom',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
             arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

plt.annotate(f'Lowest: {min_sales["Automobile_Sales"]:.0f}', 
             xy=(min_sales['Year'], min_sales['Automobile_Sales']),
             xytext=(10, -10), textcoords='offset points', ha='left', va='top',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
             arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

# Show the plot
plt.tight_layout()
plt.show()
````

````python
# Group the data by Year and Vehicle_Type, calculating the sum of Automobile_Sales
sales_by_type = df.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].sum().unstack()

# Create the line plot
plt.figure(figsize=(15, 8))
for column in sales_by_type.columns:
    plt.plot(sales_by_type.index, sales_by_type[column], label=column, marker='o')

# Customize the chart
plt.title('Automobile Sales Trends by Vehicle Type', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Automobile Sales', fontsize=12)
plt.legend(title='Vehicle Type', title_fontsize='12', fontsize='10')
plt.grid(True, linestyle='--', alpha=0.7)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Highlight recession periods
recession_years = df[df['Recession'] == 1]['Year'].unique()
for year in recession_years:
    plt.axvspan(year-0.5, year+0.5, color='gray', alpha=0.2)

# Add a text box explaining the gray areas
plt.text(0.02, 0.98, 'Gray areas indicate recession years', transform=plt.gca().transAxes, 
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# Show the plot
plt.tight_layout()
plt.show()

# Print recession years for reference
print("Recession years:", sorted(recession_years))
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart
        <br> You will need to group recession average Automobile_Sales and then plot it<br>
    Make use of sns.barplot(x=x,y=y, data = df)</p>
</details>


````python
import seaborn as sns

# Define the recession and non-recession periods
recession_period = [2008, 2009]
non_recession_period = [2012, 2013]

# Filter the data for these periods
recession_data = df[df['Year'].isin(recession_period)]
non_recession_data = df[df['Year'].isin(non_recession_period)]

# Combine the datasets and add a period column
recession_data['Period'] = 'Recession (2008-2009)'
non_recession_data['Period'] = 'Non-Recession (2012-2013)'
combined_data = pd.concat([recession_data, non_recession_data])

# Create the plot
plt.figure(figsize=(16, 8))
sns.set_style("whitegrid")
sns.set_palette("deep")

# Create the line plot
g = sns.lineplot(data=combined_data, x='Year', y='Automobile_Sales', 
                 hue='Vehicle_Type', style='Period', markers=True, 
                 dashes=False, linewidth=2.5, markersize=10)

# Customize the plot
plt.title('Comparison of Automobile Sales: Recession vs Non-Recession', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Automobile Sales', fontsize=14)
plt.legend(title='Vehicle Type & Period', title_fontsize='13', fontsize='12', loc='center left', bbox_to_anchor=(1, 0.5))

# Adjust x-axis ticks
plt.xticks(combined_data['Year'].unique())

# Add annotations for recession and non-recession periods
plt.annotate('Recession Period', xy=(2008.5, plt.ylim()[1]), xytext=(0, 10), 
             textcoords='offset points', ha='center', va='bottom',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
             fontsize=12)

plt.annotate('Non-Recession Period', xy=(2012.5, plt.ylim()[1]), xytext=(0, 10), 
             textcoords='offset points', ha='center', va='bottom',
             bbox=dict(boxstyle='round,pad=0.5', fc='lightgreen', alpha=0.5),
             fontsize=12)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
````

### Now you want to compare the sales of different vehicle types during a recession and a non-recession period
<br>We recommend that you use the functionality of **Seaborn Library** to create this visualization


<details><summary>Click here for a hint</summary>

<p>
     To visualize sales of different vehicles during recession and non-recession periods, you can use a bar chart
        <br> You will need to group Recession, Vehicle_Type for average Automobile_Sales and then plot it<br>
    Make use of sns.barplot(x=x,y=y, data = df)</p>
</details>


````python
import seaborn as sns

# Define the recession and non-recession periods
recession_period = [2008, 2009]
non_recession_period = [2012, 2013]

# Filter the data for these periods
recession_data = df[df['Year'].isin(recession_period)]
non_recession_data = df[df['Year'].isin(non_recession_period)]

# Calculate average sales for each vehicle type in each period
recession_avg = recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
recession_avg['Period'] = 'Recession (2008-2009)'

non_recession_avg = non_recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
non_recession_avg['Period'] = 'Non-Recession (2012-2013)'

# Combine the datasets
combined_data = pd.concat([recession_avg, non_recession_avg])

# Create the plot
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")
sns.set_palette("deep")

# Create the grouped bar plot
ax = sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Period', data=combined_data)

# Customize the plot
plt.title('Comparison of Average Automobile Sales by Vehicle Type: Recession vs Non-Recession', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=14)
plt.ylabel('Average Automobile Sales', fontsize=14)
plt.legend(title='Period', title_fontsize='13', fontsize='12')

# Add value labels on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import matplotlib.pyplot as plt
import seaborn as sns

# Separate recession and non-recession data
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Create the subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))
fig.suptitle('GDP Variation: Recession vs Non-Recession Periods', fontsize=16)

# Plot for Recession periods
sns.lineplot(data=recession_data, x='Year', y='GDP', ax=ax1, marker='o')
ax1.set_title('GDP During Recession Periods', fontsize=14)
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('GDP', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)

# Annotate recession periods
for year in recession_data['Year'].unique():
    ax1.axvspan(year-0.5, year+0.5, color='red', alpha=0.2)

# Plot for Non-Recession periods
sns.lineplot(data=non_recession_data, x='Year', y='GDP', ax=ax2, marker='o')
ax2.set_title('GDP During Non-Recession Periods', fontsize=14)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('GDP', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()

# Print some statistics
print("Recession Years:", sorted(recession_data['Year'].unique()))
print("Average GDP during recessions:", recession_data['GDP'].mean())
print("Average GDP during non-recessions:", non_recession_data['GDP'].mean())
````

<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for non-recession years
non_recession_data = df[df['Recession'] == 0]

# Group by Month and calculate average sales and average seasonality weight
monthly_data = non_recession_data.groupby('Month')[['Automobile_Sales', 'Seasonality_Weight']].mean().reset_index()

# Create the bubble plot
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Create scatterplot with varying bubble sizes
scatter = plt.scatter(monthly_data['Month'], 
                      monthly_data['Automobile_Sales'], 
                      s=monthly_data['Seasonality_Weight']*1000,  # Adjust size for visibility
                      alpha=0.6, 
                      c=monthly_data['Seasonality_Weight'], 
                      cmap='viridis')

# Customize the plot
plt.title('Seasonality Impact on Automobile Sales', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Automobile Sales', fontsize=12)

# Set x-axis ticks to month names
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(1, 13), month_names, rotation=45)

# Add a colorbar to show the seasonality weight
cbar = plt.colorbar(scatter)
cbar.set_label('Seasonality Weight', rotation=270, labelpad=15)

# Annotate each bubble with its month
for i, row in monthly_data.iterrows():
    plt.annotate(month_names[i], (row['Month'], row['Automobile_Sales']), 
                 xytext=(5, 5), textcoords='offset points')

# Add a text box explaining the bubble size
plt.text(0.95, 0.05, 'Bubble size represents\nseasonality weight', 
         transform=plt.gca().transAxes, fontsize=10, 
         verticalalignment='bottom', horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()

# Print some statistics
print(monthly_data.sort_values('Automobile_Sales', ascending=False))
````

 ### TASK 1.9: Develop a lineplot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period.
 <br>Analyze the effect of the unemployment rate on vehicle type and sales during the Recession Period
 #### You can create a lineplot and title the plot as 'Effect of Unemployment Rate on Vehicle Type and Sales'


<details><summary>Click here for a hint</summary>

<p>
    Filter out the data for recession period<br>
     Make use of lineplot() from seaborn and pass the relavent data</p>
</details>


````python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for recession periods
recession_data = df[df['Recession'] == 1]

# Create the line plot
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Create a line plot for each vehicle type
for vehicle_type in recession_data['Vehicle_Type'].unique():
    data = recession_data[recession_data['Vehicle_Type'] == vehicle_type]
    sns.lineplot(x='unemployment_rate', y='Automobile_Sales', data=data, label=vehicle_type, marker='o')

plt.title('Effect of Unemployment Rate on Vehicle Type and Sales during Recessions', fontsize=16)
plt.xlabel('Unemployment Rate (%)', fontsize=12)
plt.ylabel('Automobile Sales', fontsize=12)
plt.legend(title='Vehicle Type', title_fontsize='12', fontsize='10')

# Add text annotations for key points or trends
for vehicle_type in recession_data['Vehicle_Type'].unique():
    data = recession_data[recession_data['Vehicle_Type'] == vehicle_type]
    max_sales_point = data.loc[data['Automobile_Sales'].idxmax()]
    plt.annotate(f"{vehicle_type}\nMax Sales", 
                 (max_sales_point['unemployment_rate'], max_sales_point['Automobile_Sales']),
                 xytext=(5, 5), textcoords='offset points', fontsize=8, 
                 bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

plt.tight_layout()
plt.show()

# Print additional statistics
print("Correlation between Unemployment Rate and Automobile Sales during Recessions:")
for vehicle_type in recession_data['Vehicle_Type'].unique():
    data = recession_data[recession_data['Vehicle_Type'] == vehicle_type]
    correlation = data['unemployment_rate'].corr(data['Automobile_Sales'])
    print(f"{vehicle_type}: {correlation:.2f}")
````

#### You found that the datset also contains the location/city for company offices. Now you want to show the recession impact on various offices/city sales by developing a choropleth


````python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import geopandas as gpd

# Assuming df is your original dataframe
# First, let's aggregate the data by city and calculate the recession impact
city_impact = df.groupby('City')['Automobile_Sales'].agg({
    'Recession_Sales': lambda x: x[df['Recession'] == 1].mean(),
    'Non_Recession_Sales': lambda x: x[df['Recession'] == 0].mean()
})

city_impact['Recession_Impact'] = (city_impact['Recession_Sales'] - city_impact['Non_Recession_Sales']) / city_impact['Non_Recession_Sales'] * 100

# Load a GeoDataFrame with city coordinates (you might need to create this or use an external source)
# For this example, I'm assuming you have a CSV file with city coordinates
city_coords = pd.read_csv('city_coordinates.csv')  # You'll need to create or obtain this file
city_impact = city_impact.merge(city_coords, on='City')

# Create a base map centered on the mean coordinates
m = folium.Map(location=[city_impact['Latitude'].mean(), city_impact['Longitude'].mean()], zoom_start=4)

# Add a choropleth layer
folium.Choropleth(
    geo_data=city_impact,
    name='Recession Impact',
    data=city_impact,
    columns=['City', 'Recession_Impact'],
    key_on='feature.properties.City',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Recession Impact (%)'
).add_to(m)

# Add markers for each city
marker_cluster = MarkerCluster().add_to(m)

for idx, row in city_impact.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"City: {row['City']}<br>Recession Impact: {row['Recession_Impact']:.2f}%",
        tooltip=row['City']
    ).add_to(marker_cluster)

# Add layer control
folium.LayerControl().add_to(m)

# Save the map
m.save('recession_impact_map.html')

print("Map saved as 'recession_impact_map.html'. Open this file in a web browser to view the map.")
````
